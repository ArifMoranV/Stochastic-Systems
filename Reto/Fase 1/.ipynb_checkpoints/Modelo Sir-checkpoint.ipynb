```julia
function d2ydt2(y,f1,f2,f3,g1,g2,g3,h1,h2,h3)
    return [y[10],
            y[11],
            y[12],
            y[13],
            y[14],
            y[15],
            y[16],
            y[17],
            y[18],
            f1(y[1],y[2],y[3],y[4],y[5],y[6],y[7],y[8],y[9],y[10],y[11],y[12],y[13],y[14],y[15],y[16],y[17],y[18],y[19]),
            g1(y[1],y[2],y[3],y[4],y[5],y[6],y[7],y[8],y[9],y[10],y[11],y[12],y[13],y[14],y[15],y[16],y[17],y[18],y[19]),
            h1(y[1],y[2],y[3],y[4],y[5],y[6],y[7],y[8],y[9],y[10],y[11],y[12],y[13],y[14],y[15],y[16],y[17],y[18],y[19]),
            f2(y[1],y[2],y[3],y[4],y[5],y[6],y[7],y[8],y[9],y[10],y[11],y[12],y[13],y[14],y[15],y[16],y[17],y[18],y[19]),
            g2(y[1],y[2],y[3],y[4],y[5],y[6],y[7],y[8],y[9],y[10],y[11],y[12],y[13],y[14],y[15],y[16],y[17],y[18],y[19]),
            h2(y[1],y[2],y[3],y[4],y[5],y[6],y[7],y[8],y[9],y[10],y[11],y[12],y[13],y[14],y[15],y[16],y[17],y[18],y[19]),
            f3(y[1],y[2],y[3],y[4],y[5],y[6],y[7],y[8],y[9],y[10],y[11],y[12],y[13],y[14],y[15],y[16],y[17],y[18],y[19]),
            g3(y[1],y[2],y[3],y[4],y[5],y[6],y[7],y[8],y[9],y[10],y[11],y[12],y[13],y[14],y[15],y[16],y[17],y[18],y[19]),
            h3(y[1],y[2],y[3],y[4],y[5],y[6],y[7],y[8],y[9],y[10],y[11],y[12],y[13],y[14],y[15],y[16],y[17],y[18],y[19]),
            1]
end


function RK5(c1,c2,c3,dt,tf,F1,F2,F3,R)
    
    #Vectorizcion de funciones
    f1=build_function(F1[1],x1,x2,x3,y1,y2,y3,z1,z2,z3,vx1,vx2,vx3,vy1,vy2,vy3,vz1,vz2,vz3,t,expression=Val{false})
    f2=build_function(F1[2],x1,x2,x3,y1,y2,y3,z1,z2,z3,vx1,vx2,vx3,vy1,vy2,vy3,vz1,vz2,vz3,t,expression=Val{false})
    f3=build_function(F1[3],x1,x2,x3,y1,y2,y3,z1,z2,z3,vx1,vx2,vx3,vy1,vy2,vy3,vz1,vz2,vz3,t,expression=Val{false})
    g1=build_function(F2[1],x1,x2,x3,y1,y2,y3,z1,z2,z3,vx1,vx2,vx3,vy1,vy2,vy3,vz1,vz2,vz3,t,expression=Val{false})
    g2=build_function(F2[2],x1,x2,x3,y1,y2,y3,z1,z2,z3,vx1,vx2,vx3,vy1,vy2,vy3,vz1,vz2,vz3,t,expression=Val{false})
    g3=build_function(F2[3],x1,x2,x3,y1,y2,y3,z1,z2,z3,vx1,vx2,vx3,vy1,vy2,vy3,vz1,vz2,vz3,t,expression=Val{false})
    h1=build_function(F3[1],x1,x2,x3,y1,y2,y3,z1,z2,z3,vx1,vx2,vx3,vy1,vy2,vy3,vz1,vz2,vz3,t,expression=Val{false})
    h2=build_function(F3[2],x1,x2,x3,y1,y2,y3,z1,z2,z3,vx1,vx2,vx3,vy1,vy2,vy3,vz1,vz2,vz3,t,expression=Val{false})
    h3=build_function(F3[3],x1,x2,x3,y1,y2,y3,z1,z2,z3,vx1,vx2,vx3,vy1,vy2,vy3,vz1,vz2,vz3,t,expression=Val{false})
    
    #f1=sp.lambdify([x1,x2,x3,y1,y2,y3,z1,z2,z3,vx1,vx2,vx3,vy1,vy2,vy3,vz1,vz2,vz3,t],F1[0])
    ts=collect(0:dt:tf)
    TT=0#tiempo en completar orbita
    n=length(ts)
    ####################################Posiciones
    ys1=ts*0
    xs1=ts*0
    zs1=ts*0
    #
    ys2=ts*0
    xs2=ts*0
    zs2=ts*0
    #
    ys3=ts*0
    xs3=ts*0
    zs3=ts*0
    
    #####################################Velocidades
    vys1=ts*0
    vxs1=ts*0
    vzs1=ts*0
    #
    vys2=ts*0
    vxs2=ts*0
    vzs2=ts*0
    #
    vys3=ts*0
    vxs3=ts*0
    vzs3=ts*0
    ###########################3##aceleraciones
    ays1=ts*0
    axs1=ts*0
    azs1=ts*0
    #
    ays2=ts*0
    axs2=ts*0
    azs2=ts*0
    #
    ays3=ts*0
    axs3=ts*0
    azs3=ts*0;
#############Condiciones iniciales
    xs1[1]=c1[1]
    ys1[1]=c1[2]
    zs1[1]=c1[3]
    vxs1[1]=c1[4]
    vys1[1]=c1[5]
    vzs1[1]=c1[6]
    #############
    xs2[1]=c2[1]
    ys2[1]=c2[2]
    zs2[1]=c2[3]
    vxs2[1]=c2[4]
    vys2[1]=c2[5]
    vzs2[1]=c2[6]
    ###########
    xs3[1]=c3[1]
    ys3[1]=c3[2]
    zs3[1]=c3[3]
    vxs3[1]=c3[4]
    vys3[1]=c3[5]
    vzs3[1]=c3[6];
    for i in 1:n-1
        #Runge Kutta
        z0=[xs1[i],xs2[i],xs3[i],ys1[i],ys2[i],ys3[i],zs1[i],zs2[i],zs3[i],vxs1[i],vxs2[i],vxs3[i],vys1[i],vys2[i],vys3[i],vzs1[i],vzs2[i],vzs3[i],ts[i]]
        k1=d2ydt2(z0,f1,f2,f3,g1,g2,g3,h1,h2,h3)
        k2=d2ydt2(z0+(dt*k1)/2,f1,f2,f3,g1,g2,g3,h1,h2,h3)
        k3=d2ydt2(z0+dt*(3*k1+k2)/16,f1,f2,f3,g1,g2,g3,h1,h2,h3)
        k4=d2ydt2(z0+dt*k3/2,f1,f2,f3,g1,g2,g3,h1,h2,h3)
        k5=d2ydt2(z0+dt*(-3*k2+6*k3+9*k4)/16,f1,f2,f3,g1,g2,g3,h1,h2,h3)
        k6=d2ydt2(z0+dt*(k1+4*k2+6*k3-12*k4+8*k5)/7,f1,f2,f3,g1,g2,g3,h1,h2,h3)
        
        k1=7*(k1+k6)
        k3=16*k3
        k2=16*k5
        k4=12*k4
        
        
        #Vectores M1
        xs1[i+1]=xs1[i]+(dt/90)*(k1[1]+2*k2[1]+2*k3[1]+k4[1])
        ys1[i+1]=ys1[i]+(dt/90)*(k1[4]+2*k2[4]+2*k3[4]+k4[4])
        zs1[i+1]=zs1[i]+(dt/90)*(k1[7]+2*k2[7]+2*k3[7]+k4[7])
        vxs1[i+1]=vxs1[i]+(dt/90)*(k1[10]+2*k2[10]+2*k3[10]+k4[10])
        vys1[i+1]=vys1[i]+(dt/90)*(k1[13]+2*k2[13]+2*k3[13]+k4[13])
        vzs1[i+1]=vzs1[i]+(dt/90)*(k1[16]+2*k2[16]+2*k3[16]+k4[16])
        
        axs1[i]=f1(xs1[i],xs2[i],xs3[i],ys1[i],ys2[i],ys3[i],zs1[i],zs2[i],zs3[i],vxs1[i],vxs2[i],vxs3[i],vys1[i],vys2[i],vys3[i],vzs1[i],vzs2[i],vzs3[i],ts[i])    
        ays1[i]=f2(xs1[i],xs2[i],xs3[i],ys1[i],ys2[i],ys3[i],zs1[i],zs2[i],zs3[i],vxs1[i],vxs2[i],vxs3[i],vys1[i],vys2[i],vys3[i],vzs1[i],vzs2[i],vzs3[i],ts[i])
        azs1[i]=f3(xs1[i],xs2[i],xs3[i],ys1[i],ys2[i],ys3[i],zs1[i],zs2[i],zs3[i],vxs1[i],vxs2[i],vxs3[i],vys1[i],vys2[i],vys3[i],vzs1[i],vzs2[i],vzs3[i],ts[i])
        
        #################Vectores M2
        xs2[i+1]=xs2[i]+(dt/90)*(k1[2]+2*k2[2]+2*k3[2]+k4[2])
        ys2[i+1]=ys2[i]+(dt/90)*(k1[5]+2*k2[5]+2*k3[5]+k4[5])
        zs2[i+1]=zs2[i]+(dt/90)*(k1[8]+2*k2[8]+2*k3[8]+k4[8])
        vxs2[i+1]=vxs2[i]+(dt/90)*(k1[11]+2*k2[11]+2*k3[11]+k4[11])
        vys2[i+1]=vys2[i]+(dt/90)*(k1[14]+2*k2[14]+2*k3[14]+k4[14])
        vzs2[i+1]=vzs2[i]+(dt/90)*(k1[17]+2*k2[17]+2*k3[17]+k4[17])
         
        axs2[i]=g1(xs1[i],xs2[i],xs3[i],ys1[i],ys2[i],ys3[i],zs1[i],zs2[i],zs3[i],vxs1[i],vxs2[i],vxs3[i],vys1[i],vys2[i],vys3[i],vzs1[i],vzs2[i],vzs3[i],ts[i])    
        ays2[i]=g2(xs1[i],xs2[i],xs3[i],ys1[i],ys2[i],ys3[i],zs1[i],zs2[i],zs3[i],vxs1[i],vxs2[i],vxs3[i],vys1[i],vys2[i],vys3[i],vzs1[i],vzs2[i],vzs3[i],ts[i])
        azs2[i]=g3(xs1[i],xs2[i],xs3[i],ys1[i],ys2[i],ys3[i],zs1[i],zs2[i],zs3[i],vxs1[i],vxs2[i],vxs3[i],vys1[i],vys2[i],vys3[i],vzs1[i],vzs2[i],vzs3[i],ts[i])
        
        ###Vectores M3############################################
        
        xs3[i+1]=xs3[i]+(dt/90)*(k1[3]+2*k2[3]+2*k3[3]+k4[3])
        ys3[i+1]=ys3[i]+(dt/90)*(k1[6]+2*k2[6]+2*k3[6]+k4[6])
        zs3[i+1]=zs3[i]+(dt/90)*(k1[9]+2*k2[9]+2*k3[9]+k4[9])
        vxs3[i+1]=vxs3[i]+(dt/90)*(k1[12]+2*k2[12]+2*k3[12]+k4[12])
        vys3[i+1]=vys3[i]+(dt/90)*(k1[15]+2*k2[15]+2*k3[15]+k4[15])
        vzs3[i+1]=vzs3[i]+(dt/90)*(k1[18]+2*k2[18]+2*k3[18]+k4[18])
         
        axs3[i]=h1(xs1[i],xs2[i],xs3[i],ys1[i],ys2[i],ys3[i],zs1[i],zs2[i],zs3[i],vxs1[i],vxs2[i],vxs3[i],vys1[i],vys2[i],vys3[i],vzs1[i],vzs2[i],vzs3[i],ts[i])    
        ays3[i]=h2(xs1[i],xs2[i],xs3[i],ys1[i],ys2[i],ys3[i],zs1[i],zs2[i],zs3[i],vxs1[i],vxs2[i],vxs3[i],vys1[i],vys2[i],vys3[i],vzs1[i],vzs2[i],vzs3[i],ts[i])
        azs3[i]=h3(xs1[i],xs2[i],xs3[i],ys1[i],ys2[i],ys3[i],zs1[i],zs2[i],zs3[i],vxs1[i],vxs2[i],vxs3[i],vys1[i],vys2[i],vys3[i],vzs1[i],vzs2[i],vzs3[i],ts[i]);
    end
    nn=Int(round(n/1000))
    if R==true
        
        gr(size=(800,500))
        plot(xs1[1:nn:end],ys1[1:nn:end],zs1[1:nn:end],label="masa 1",color="yellow",lw=5)
        plot!(xs2[1:nn:end],ys2[1:nn:end],zs2[1:nn:end],label="masa 2",color="blue",lw=2)
        plot!(xs3[1:nn:end],ys3[1:nn:end],zs3[1:nn:end],label="masa 3",color="red",lw=2)       
    end
    
    return (xs1[1:nn:end],xs2[1:nn:end],xs3[1:nn:end],ys1[1:nn:end],ys2[1:nn:end],ys3[1:nn:end],zs1[1:nn:end],zs2[1:nn:end],zs3[1:nn:end],vxs1[1:nn:end],vxs2[1:nn:end],vxs3[1:nn:end],vys1[1:nn:end],vys2[1:nn:end],vys3[1:nn:end],vzs1[1:nn:end],vzs2[1:nn:end],vzs3[1:nn:end],axs1[1:nn:end],axs2[1:nn:end],axs3[1:nn:end],ays1[1:nn:end],ays2[1:nn:end],ays3[1:nn:end],azs1[1:nn:end],azs2[1:nn:end],azs3[1:nn:end],ts[1:nn:end]);
    ########Graficas####################################
end  
```

```julia
n=length(ts)
mm=5
xx=vcat(xs1,xs2,xs3)
yy=vcat(ys1,ys2,ys3)
sc=1/100
anim2d=@animate for i in 1:10:n;
    gr(size=(1000,1000))
    Legend=(bgcolor=(:dark))
    a=1
    j=1
    if i>1
        j=Int(ceil(i*0.5))
        a=range(0,1,length=(i-j+1))
    end 
    Plots.plot(xs1[j:i],ys1[j:i],color="yellow",lw=2,xlim = [minimum(xx)-1, maximum(xx)+1],
    ylim = [minimum(yy)-1, maximum(yy)+1],label="",xlabel="x",ylabel="y",bgcolor="black",grid=false,axis=false,alpha=a);
    Plots.scatter!([xs1[i]],[ys1[i]], color = "yellow", markersize = mm,label="Masa1");
    Plots.quiver!([xs1[i]],[ys1[i]],quiver=([axs1[i].*sc],[ays1[i].*sc]),color="yellow",lw=3,length=1/80)
    
    Plots.plot!(xs2[j:i],ys2[j:i],color="blue",lw=2,label="",alpha=a);
    Plots.scatter!([xs2[i]],[ys2[i]], color = "blue", markersize = mm ,label="Masa2");
    Plots.quiver!([xs2[i]],[ys2[i]],quiver=([axs2[i].*sc],[ays2[i].*sc]),color="blue",lw=3,length=1/80)
    
    Plots.plot!(xs3[j:i],ys3[j:i],color="red",lw=2,label="",alpha=a);
    Plots.scatter!([xs3[i]],[ys3[i]], color = "red", markersize = mm,label="Masa3");
    Plots.quiver!([xs3[i]],[ys3[i]],quiver=([axs3[i].*sc],[ays3[i].*sc]),color="red",lw=3,length=1/80,normalize=true)
    #
end;
```
