```julia
using Plots
using Random
using Distributions
using LinearAlgebra
```

```julia
N=30
rs=rand(N,3)
n=1000# Iteraciones
```

```julia
##Graficar


Δr=vcat(rs,rs[1,:]')##Trayectoria
D=abs.(diff(Δr,dims=1));
Ds=norm(sum(D,dims=1));


#Dss=zeros(n)
Dss=Ds
Data_ty=Dict();
Data_ty[1]=rs;
Ds
###########################
#plot!(Δr[:,1],Δr[:,2],Δr[:,3])
#display(p)
```

```julia
Dss=Ds
for i in 1:n
    rss=rs[shuffle(1:end), :]
    dd,Δr2=dista(rss)
    if dd<Dss
        Data_ty[1]=rss
        Data_ty[2]=Δr2
        Dss=dd
    end
    return Dss,rss,Data_ty
end
Dss
```

```julia
gr(size=(500,500))
p = plot()#
####Ciudades
for i in 1:N
    scatter!([rs[i,1]],[rs[i,2]],[rs[i,3]],label="")
end
plot!(Δr[:,1],Δr[:,2],Δr[:,3],label="Distancia Inicial="*string(Ds),color="red")
plot!(Data_ty[2][:,1],Data_ty[2][:,2],Data_ty[2][:,3],color="green",label="Distancia Optima="*string(Dss))

#display(p)
```

```julia
function dista(rs)
    Δr=vcat(rs,rs[1,:]')##Trayectoria
    D=abs.(diff(Δr,dims=1))
    Ds=norm(sum(D,dims=1))
    return Ds,Δr
end
#plot!(Δr[:,1],Δr[:,2],Δr[:,3])
#display(p)
```
