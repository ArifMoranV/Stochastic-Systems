```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=10000
data=np.random.randint(0,2,N)
intdata=data*0
for i in range(0,N):
        intdata[i]=sum(data[0:i+1])
plt.plot(data)
```

```python
plt.plot(intdata,color='red',label='Integral')
plt.grid('on')
```

## Promedio

```python
av=np.zeros(N)
for i in range(0,N):
    av[i]=np.mean(data[0:i])
plt.plot(av,label='N='+str(N))
plt.grid('on')
```

## Colonia 2

```python
N=10000
data=np.ones(N)
Int=np.zeros(N)

for i in range(0,N):
    rn=np.random.randint(0,2)
    j=1
    while rn==0:
        rn=np.random.randint(0,2)
        j+=1
    data[i]=2**j
    Int[i]=sum(data[0:i])

        
plt.plot(data,label='Bacterias')
plt.plot(Int,label='Integral')
plt.grid('on')
plt.legend()
```

## Promedio

```python
av=np.zeros(N)
for i in range(0,N):
    av[i]=np.mean(data[0:i])
plt.plot(av,label='N='+str(N))
plt.legend()
plt.grid('on')
```
