En clase definimos el proceso de Wiener como:
$$W(t)=\int_0^tdW$$

donde dW es una variable aleatoria con distribuci√≥n Gausiana, $\langle dW \rangle=0 $ y $\langle dW^2\rangle =dt$

Simulando el proceso de Wiener num√©ricamente en un intervalo de 100 segundos. Es decir, Dividiremos el espacio del tiempo en $ùëÅ$ intervalos $ùëëùë°$.

```python
import numpy as np
import random
from matplotlib import pyplot as plt
```

1. Utiliza un generador de n√∫meros aleatorios para simular los incrementos $dW$.Asegurandose que la varianza de $dW$ es igual a $dt$ y grafica $dW$

```python
def weiner(N,tf,qq):
    dt=tf/N
    ts=np.linspace(0,tf,N)
    if qq==1:
        variance=np.sqrt(dt)
    else:
        variance=dt
    dW=np.random.normal(0, variance, size=N)
    dW[0]=0
    
    W=np.ones(N)
    ########################Integral####################
    for i in range(0,N):
        W[i]=sum(dW[0:i+1])
    return ts,dW,W
```

```python
ts,dW,W=weiner(1000,100,1)
plt.plot(ts,dW,label='$dW$ vs $t$')
plt.legend()
```

2. Realiza una suma acumulada de los incrementos para obtener los valores del proceso $W(t)$ como funci√≥n del tiempo $t$. Grafica $W(t)$ vs $t$

## Integral $W(t)$
$$W(t)=\int_{0}^t{dW}$$

$$\sigma_(dW)^2=dt$$

```python
plt.plot(ts,W,label='$W$ vs $t$')
plt.legend()
```

```python
def its(N,tf,NN,graf,qq):###Varias iteraciones el experimento de Weiner
    M=np.zeros((NN,N))#Base datos
    M2=M*0
    for i in range(0,NN):
        ts,dW,W=weiner(N,tf,qq);
        M[i,:]=W
        M2[i,:]=dW
    T,Y=np.meshgrid(ts,np.zeros(NN))
    fig1=plt.figure()
    if graf==True:
        for j in  range(NN):
            plt.plot(T[j,:],M[j,:],color='blue',lw=0.2,alpha=0.3,label='$W$ vs $t$')
        #plt.legend()
        plt.grid('on')
        plt.show()
    return fig1,M,M2,T 
```

3. Realiza el proceso un m√≠nimo de 300 veces. Muestra en una sola gr√°fica todas las realizaciones de $ùëä(ùë°)$

```python
N=200
tf=100
NN=300 ### Numero de experimentos
graf=True
fig,Ws,dWs,T=its(N,tf,NN,graf,1)
Matrix[1]=Ws
Matrix[2]=T
```


## Varianza
4.Grafica la varianza $W(ùë°) $, $ \sigma^2 \langle ùëä(ùë°)\rangle=\langle W(t)^2 \rangle-\langle W(t)\rangle^2 $ conforme pasa el
tiempo. ¬øEs lo que esperas obtener?
$$\sigma^2(W(t))=\langle W(t)^2 \rangle-\langle W(t) \rangle^2$$

```python
def vari(M,NN,graf,ts):
    var=np.sum(M**2,0)/NN
    mean=(np.sum(M,0)/NN)**2
    sigma=var-mean
    if graf==True:
        fig=plt.figure()
        plt.plot(ts,sigma,color='black',label='$\sigma^2$ vs $t$, '+'Num Exps='+str(NN))
        plt.grid('on')
        plt.axis('equal')
        plt.legend()
    return fig,sigma
        
```

```python
fig,v=vari(Ws,NN,True,T[0,:])
```

7. Demuestra anal√≠ticamente que s√≥lo cuando $ùúé^2(ùëëùëä) = ùëëùë°$, la suma de los intervalos
$\sum_0^N ùëëùëä$ es independiente del n√∫mero de intervalos.
En clase vimos que un proceso $ùë•(ùë°)$ descrito por la ecuaci√≥n diferencial estoc√°stica:

$$dx=f(x,t)dt+g(x,t)dW$$

Tiene una distribuci√≥n de probabilidad $ùëÉ(ùë•, ùë°)$ dada por la ecuaci√≥n de Fokker-Planck:
$$\frac{\partial P}{\partial t}=-\frac{\partial [P f]}{\partial x}+\frac{g^2}{2}\frac{\partial^2 P}{\partial x^2}$$

$$\sigma^2(x(t))=\sum_{i=0}^N{\Delta t}^\gamma=N{\Delta t}^\gamma$$

$$\Delta t=(\frac{t}{N})^\gamma\therefore \sigma^2(x(t))= t^\gamma\frac{N}{N^\gamma}=t^\gamma N^{1-\gamma}$$
Para que el proceso de Wienersirva para describir un sistema f√≠sico, es necesario que $ùõæ$ = 1

$$\sigma^2(x(t))=t$$

### $$P(x,t_0)=\frac{1}{\sigma\sqrt{2\pi}}\exp^{\frac{(x-\mu)^2}{2\sigma^2}}$$

```python
import sympy as sp
```

```python
M=Matrix[5]
var=np.sum(M**2,0)/NN
mean=(np.sum(M,0)/NN)**2
sigma=var-mean
```

```python
x,u,s=sp.symbols('x,mu,sigma')
```

```python
f=(1/(s*sp.sqrt(2*sp.pi)))*sp.exp(-1/2*((x-u)/s)**2)
f
```

```python
f1=sp.lambdify([x,u,s],f)
```

```python
xs=np.linspace(-100,100,len(sigma))
```

```python
dd=np.array([25.,50.,75.,100.])[::-1]
for i in range(len(dd)):
    j=5
    #figure=plt.figure()
    #plt.grid('on')
    plt.hist((Matrix[j][:,np.where(np.round(Matrix[j+1][0,:],0)==dd[i])[0][0]]),label="$\langle dW^2 ‚ü© = dt^2$, $t=$"+str(dd[i]),alpha=0.4)
    plt.plot(xs,500*f1(xs,0,np.sqrt(sigma[np.where(np.round(Matrix[j+1][0,:],0)==dd[i])[0][0]])),alpha=0.5,label='Solucion Analitica')
    #plt.axis('equal')
    plt.legend()
```
