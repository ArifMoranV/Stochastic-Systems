La ecuación de Langevin surge de agregar a la ecuación de movimiento de Newton en un medio con arrastre, un término que representa las fluctuaciones térmicas. En ausencia de otras fuerzas externas, esta ecuación es:
$$ma=-\gamma v+\eta$$

donde $m$ es la masa, $a$ es la aceleración, $v$ es la velocidad, 
 es el coeficiente de arrastre, y 
 $\gamma$ es una variable aleatoria que representa el ruido térmico, con $\langle \eta  \rangle=0$
 y $\langle \eta^2\rangle=2\gamma k_B T$


$$m\frac{dv}{dt}=-\gamma v+\eta$$
no es un proceso diferenciable , por lo tanto queremos replantearla

$$\frac{dv}{dt}=-\frac{\gamma v}{m}+\frac{\eta}{m}$$

$${\eta(x,t) dt} \rightarrow g(x,t) dW$$

$$dv=-\frac{\gamma v}{m}dt+ \frac{g}{m} dW$$

si sabemos que $\langle \eta \rangle=0$

$$dv=-\frac{\gamma v}{m}dt$$

$$v(t+dt)=v(t)-\frac{\gamma v}{m}dt$$

$$v(t+dt)=v(t)[1-\frac{\gamma v}{m}dt]+\frac{g}{m} dW$$

$$\langle v(t+dt) \rangle=\langle v(t)[1-\frac{\gamma v}{m}dt]\rangle+\langle \frac{g}{m} dW\rangle$$

$$\langle dW \rangle=0$$

$$\langle v(t+dt) \rangle=\langle v(t)[1-\frac{\gamma v}{m}dt]\rangle$$

Expandiendo en serie de taylor:

$$ e^{-\frac{\gamma}{m}dt}\approx 1-\frac{\gamma}{m}dt+\frac{1}{2}\frac{\gamma^2}{m^2}dt^2\mp...$$

truncando la serie:

$$\langle v(t+dt) \rangle=\langle v(t) \rangle e^{-\frac{\gamma}{m}dt} $$


$$t=0 \therefore \langle v(0+dt) \rangle=v_0 e^{-\frac{\gamma}{m}dt}$$

$$t=Ndt; v(Ndt+dt) \rangle=v_0 e^{-\frac{\gamma}{m}Ndt}$$

### $$\boxed{\langle v(t) \rangle=v_0 e^{-\frac{\gamma}{m}t}}$$


### Para calcular la Varianza 

$${\sigma(v)}^2=\langle v^2 \rangle -\langle v \rangle^2$$
utilizando el calculo de Ito


$$d(v^2)=\frac{d(v^2)}{dv}dv+\frac{1}{2}\frac{d^2(v^2)}{d{v}^2}dv^2$$

$$d(v^2)=2v dv +dv^2$$

sustituyendo $dv$:


$$d(v^2)=-2v^2\frac{\gamma}{m}dt +2v\frac{g}{m}dW+[\frac{\gamma^2}{m^2}dt^2-2\frac{\gamma^3}{m^3}v^3dt^3+\frac{g^2}{m^2}dW^2]$$

$$$$

$$d(v^2)=-2v^2\frac{\gamma}{m}dt +2v\frac{g}{m}dW+\frac{g^2}{m^2}dt \iff dW^2=dt , dt=0\times $$

$$\frac{d\langle v^2 \rangle}{dt}=-2\frac{\gamma}{m}\langle v^2 \rangle+\frac{g^2}{m^2}$$

resolviendo la ecuacion:

$$\langle v^2 \rangle =(v_0-\frac{g^2}{2m\gamma})e^{\frac{2\gamma}{m}t}+\frac{g^2}{2m\gamma}$$

$${\sigma(v)}^2=\langle v^2 \rangle-\langle v\rangle^2$$

### $$\boxed{\sigma^2(v)=(1-e^{-\frac{2\gamma t}{m}})\frac{g^2}{2m\gamma}}$$


para que eta este ralacionada con la temperatura  g debe cumplir on la relacion dada pro $$g=\sqrt{2k_B T\gamma}$$

```julia
using Plots
using Distributions
using Random
```

```julia
T=300 #Kelvin
tf=1000
n=2000# numero de datos en el tiempo
N=30# Numero de realizaciones
ts=range(0,tf,length=n)
dt=tf/n
v0=10 #μm s-1
m=0.5 #
γ=0.01 #pN s um-1
β=1
var=dt^2
```

```julia
vt=v0*exp.(-γ*ts./m)
plot(ts,vt)
```

```julia
function f(x,dw,dt)
    dv=-γ*x*dt/m+dw/m
end
```

```julia
function pepe(n,dt,v0,var)
    xs=zeros(n);
    vs=zeros(n);
    vs[1]=v0
    ax=zeros(n);
    dW=rand(Normal(0, sqrt(var)), n)
    for i in 1:n-1
        vs[i+1]=vs[i]+f(vs[i],dW[i],dt)
        xs[i+1]=xs[i]+vs[i]*dt
    end
    return(vs)
end
```

```julia
function its(N,n,dt,v0,var)
    M=zeros(N,n);#Matriz con las diferentes realizaciones Nxn
    for i in 1:N
        M[i,:]=pepe(n,dt,v0,var,)
    end
    return M
end
```

```julia
T = repeat(ts, 1, N)'
MM=its(N,n,dt,v0,var)
plot(T[:,:],MM[:,:],color="black",lw=1/20,label=false,alpha=0.3)
plot!(ts,vt,label=" v(t) Analitico")
```
