```julia
using Plots
```

```julia

n=10000
N=3
```

```julia
PI=[0.75 0.25 0; 0 0.9 0.1; 0.05 0 0.95]'
#PI2=[0.5 0.25 0; 0 0.5 0.1; 0.05 0 0.5]'
```

```julia
SIR=zeros(N,n);
SIR[:,1]=[0.99,0.01,0]#Probabilidad inicial
for i in 1:n-1
    SIR[:,i+1]=PI*SIR[:,i]
end
```

```julia
SIR
```

```julia
plot(SIR[1,1:200])
plot!(SIR[2,1:200])
plot!(SIR[3,1:200])
```
