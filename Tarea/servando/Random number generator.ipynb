```julia
using Plots
```

```julia
N=10000
x0=6
xs=zeros(N);## semilla
xs[1]=x0
a=2^16+3
c=0
m=2^31
```

```julia
M=zeros(3,N);
```

```julia
for i in 1:N-1
    xs[i+1]=(a*(xs[i])+c)%m
end
```

```julia

M[3,:]=xs;
```

```julia
scatter(M[1,:]/m,M[2,:]/m,M[3,:]/m,markersize = 1)
```

## Creacion de n'umros aleatorios uniformemente distribuidos entre a y b
$$r=a+(b-a)u$$

```julia
a=2;
y1=-(1/a)*log.(M[1,:]/m);
```

```julia
histogram(y1,label="gay",color="red")
```

```julia

y2=sqrt.(-2*log.(M[1,:]/m)).*cos.(2*pi.*M[3,:]/m);
histogram(y2)
#scatter(y2)
```
